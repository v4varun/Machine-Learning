```python
def chat_with_model(model_id, system_message, user_request):
    try:
        response = client.chat.completions.create(
            model = model_id,
            messages = [
                {"role": "system", "content": system_message},
                {"role": "user", "content": user_request}
            ]
        )

        chat_response = response.choices[0].message.content

        return chat_response

    except Exception as e:
        return str(e)
```

```python
model_id = 'gpt-3.5-turbo'

system_message = """
    As a movie recommender who specializes in various
    genres, provide information about highly-rated
    and critically acclaimed movies in the following format:
    Title:
    Director:
    Release year:
    Genre:
    Plot summary:
    Cast:
    IMDb rating:
    Rotten Tomatoes rating:
    Why I recommend it:
"""

user_request = "Could you recommend a fun movie for me to watch?"

response = chat_with_model(model_id, user_request, system_message)

print(response)
```

#### Zero-shot prompting

```python
def zero_shot(model_id, user_message):
    try:
        response = client.chat.completions.create(
            model = model_id,
            messages = [
                {"role": "user", "content": user_message}
            ]
        )

        chat_response = response.choices[0].message.content

        return chat_response

    except Exception as e:
        return str(e)
```

#### observe we can also add constraints as from which particular channel we need the movie from

```python
model_id = "gpt-3.5-turbo"

user_message = """
    If a wizard were to recommend a movie to a young apprentice, what movie would they recommend?
    constraints: Cartoon Network
"""

response = zero_shot(model_id, user_message)

print(response)
```

```python
model_id = "gpt-3.5-turbo"

user_message = """
    If a wizard were to recommend a movie to a young apprentice, what movie would they recommend?
    constraints: Disney
"""

response = zero_shot(model_id, user_message)

print(response)
```

### Streaming

The OpenAI API provides the ability to stream responses back to a client in order to allow partial results for certain requests. To achieve this, we follow the Server-sent events standard.

https://html.spec.whatwg.org/multipage/server-sent-events.html#server-sent-events

Should be able to see the content streaming in

```python
stream = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {
            "role": "user",
            "content": "Write an email explaining to your boss why your project will be delayed"
        }
    ],
    stream=True,
)

for chunk in stream:
    if chunk.choices[0].delta.content is not None:
        print(chunk.choices[0].delta.content, end="")
```
